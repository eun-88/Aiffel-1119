```python
import random

class Account:
    #생성된 계좌 개수 저장
    account_count = 0

    def __init__(self, name, balance):
        #은행이름
        self.bank = "모두은행"

        #예금주
        self.name = name

        # 계좌번호 생성 (3자리-2자리-6자리)
        num1 = str(random.randint(0, 999))
        num2 = str(random.randint(0, 99))
        num3 = str(random.randint(0, 999999))
        self.account_number = f"{num1}-{num2}-{num3}"

        #잔액
        self.balance = balance

        # 입출금 내역 저장 리스트
        self.cycle = 0
        self.deposit_log = []
        self.withdraw_log = []

        # 생성된 계좌 개수 증가
        Account.account_count += 1

    # 입금 로직
    def deposit(self, deposit_amount) :
        # 1원 이상인지 확인
        if deposit_amount < 1 :
            print("입금은 1원 이상만 가능합니다.")
            return

        self.balance += deposit_amount
        self.cycle += 1
        self.deposit_log.append(deposit_amount)

        print(f'{self.cycle}회: 입금 \t금액: {deposit_amount:,} \t잔액: {self.balance:,}')

        # 5번 이상 입금이면 이자 지급
        if len(self.deposit_log) % 5 == 0:
            interest = self.balance * 0.01
            self.balance += interest
            self.cycle += 1
            print(f'{self.cycle}회: 이자지급 \t금액: {interest:,} \t잔액: {self.balance:,}')

        if (len(self.deposit_log) + len(self.withdraw_log)) % 10 == 0:
            print('-' * 60)

    # 출금 로직
    def withdraw(self, withdraw_amount) :
        # 잔액이 충분한지 확인
        if withdraw_amount > self.balance :
            print('잔액이 부족합니다.')
            return

        self.balance -= withdraw_amount
        self.cycle += 1
        self.withdraw_log.append(withdraw_amount)

        print(f'{self.cycle}회: 출금 \t금액: {withdraw_amount:,} \t잔액: {self.balance:,}')

        if (len(self.deposit_log) + len(self.withdraw_log)) % 10 == 0:
            print('-' * 60)

    # 클래스 메소드
    @classmethod
    def get_account_num(cls) :
        print(f'지금까지 생성된 계좌 계수는 {cls.account_count}개 입니다.')

    # 히스토리 출력
    def history(self):
      print(f'입금 내역: {self.deposit_log}')
      print(f'출금 내역: {self.withdraw_log}')
```
